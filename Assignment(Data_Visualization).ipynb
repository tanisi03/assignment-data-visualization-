```python
#Q5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.
```

```python
np.random.seed(0)
data_size = 100
x = np.random.rand(data_size) * 100
y = x + (np.random.rand(data_size) * 20)
z = y + (np.random.rand(data_size) * 10)

df = pd.DataFrame({'X': x, 'Y': y, 'Z': z})
correlation_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
fig_surface = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig_surface.update_layout(title='3D Surface Plot of z=sin(x+y)',
                          scene=dict(xaxis_title='X',
                                     yaxis_title='Y',
                                     zaxis_title='Z'))
fig_surface.show()
```

BOKEH ASSIGNMENT:

```python
#Q1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

output_notebook()

x = np.linspace(0, 10, 100)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')
p.line(x, y, legend_label="Sine Wave", line_width=2)

show(p)
```

```python
#Q2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.randint(10, 100, size=n)
colors = np.random.choice(['red', 'green', 'blue', 'orange', 'purple'], size=n)
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.scatter('x', 'y', size='sizes', color='colors', source=data, legend_field='colors', fill_alpha=0.6)
show(p)
```

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
p = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency')
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],line_color="white", fill_color="blue", fill_alpha=0.7)
show(p)
```
